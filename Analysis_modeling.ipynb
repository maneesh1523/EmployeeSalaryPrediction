```python

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import joblib
```

```python
# You can replace this with your actual dataset if you have one
data = pd.DataFrame({
    'years_at_company': [0, 1, 2, 3, 4, 5, 6, 7, 8],
    'job_rate': [1.0, 2.5, 3.0, 3.5, 4.0, 3.8, 4.5, 4.8, 5.0],
    'salary': [25000, 32000, 40000, 45000, 52000, 58000, 60000, 68000, 72000]
})
```

```python
X = data[['years_at_company', 'job_rate']]
y = data['salary']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
joblib.dump(model, "linearmodel.pkl")
```
